```python
df_col = ['conversation_id', 'date','tweet', 'language','username', 'nlikes', 'nreplies', 'nretweets']   

file_10 = pd.read_csv('bitcoin_10.csv',engine = 'python')[df_col]
file_13 = pd.read_csv('bitcoin_13.csv',engine = 'python')[df_col]
file_16 = pd.read_csv('bitcoin_16.csv',engine = 'python')[df_col]
```

```python
df = pd.concat([file_10,file_13,file_16], axis = 0)
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df = df.dropna().sort_values(by = 'date')


df = df[df['language'] != 'und']
df = df[df['language'] != 'qme']
df = df[df['language'] != 'qht']
df = df[df['language'] != 'qam']
df = df[df['language'] != 'zxx']
df = df[df['language'] != 'hy']
df = df[df['language'] != 'te']
df = df[df['language'] != 'lv']
df = df[df['language'] != 'ta']
df = df[df['language'] != 'am']
df = df[df['language'] != 'my']
df = df[df['language'] != 'ne']
df = df[df['language'] != 'ta']
df = df[df['language'] != 'he']
df = df[df['language'] != 'pa']
df = df[df['language'] != 'ug']
df = df[df['language'] != 'iw']


df = df.replace({'language':lang_dict})

del file_10,file_13,file_16
```

```python
merged_prev = pd.read_csv('bitcoin_merged.csv',index_col=False, engine = 'python').drop('Unnamed: 0',axis = 1)
merged_prev['date'] = pd.to_datetime(merged_prev['date'], errors='coerce')

```

```python
df_new = pd.concat([df,merged_prev], axis = 0).drop_duplicates()
df_new['date'] = pd.to_datetime(df_new['date'], errors='coerce')
df = df_new.sort_values(by = 'date')
del df_new, merged_prev

```

```python
df.to_csv('new_df_btc.csv')
```
